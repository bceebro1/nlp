```python
from transformers import pipeline
```

```python
classifier = pipeline(
    task="sentiment-analysis",
    model="distilbert-base-uncased-finetuned-sst-2-english"
)
```

```python
text = "I absolutely loved this movie. It was amazing!"

result = classifier(text)

print("Input Text:", text)
print("Sentiment:", result[0]["label"])
print("Confidence:", round(result[0]["score"], 4))
```

```python
texts = [
    "I love this phone.",
    "The service was terrible.",
    "The food was okay.",
    "I will definitely buy it again!",
    "This is the worst experience of my life."
]

results = classifier(texts)

for text, result in zip(texts, results):
    print(f"Text: {text}")
    print(f"Sentiment: {result['label']}")
    print(f"Confidence: {result['score']:.4f}")
    print("-" * 40)
```

```python
user_text = input("Enter a sentence: ")

prediction = classifier(user_text)

print("\nPrediction")
print("Sentiment:", prediction[0]["label"])
print("Confidence:", prediction[0]["score"])
```

```python
# Install required libraries (Run this only once in Colab)
!pip install -q transformers sentencepiece torch

# Import libraries
from transformers import T5ForConditionalGeneration, T5Tokenizer

# Load the pre-trained T5 model and tokenizer
model_name = "t5-small"

tokenizer = T5Tokenizer.from_pretrained(model_name)
model = T5ForConditionalGeneration.from_pretrained(model_name)

# Input text
text = """
Nepal is a beautiful landlocked country located in South Asia, situated between China to the north and India to the south, east, and west. It is widely known for its breathtaking Himalayan mountain range, including Mount Everest, the highest peak in the world. Nepal is rich in natural beauty, cultural diversity, and historical heritage. The country is home to hundreds of ethnic groups, each with its own language, traditions, festivals, and customs, making it one of the most culturally diverse nations in the region. Kathmandu, the capital city, is famous for its ancient temples, UNESCO World Heritage Sites, vibrant markets, and religious significance to both Hindus and Buddhists. Tourism plays a major role in Nepal's economy, attracting visitors from around the world who come for trekking, mountaineering, wildlife safaris, adventure sports, and spiritual experiences. Popular destinations include Pokhara, Chitwan National Park, Lumbini, the birthplace of Lord Buddha, and the Everest and Annapurna regions. Agriculture remains an important occupation for a large portion of the population, with rice, maize, wheat, and millet being among the major crops grown in the fertile plains and hills. In recent years, Nepal has also made progress in areas such as education, information technology, renewable energy, and infrastructure development, although challenges such as poverty, unemployment, and natural disasters still exist. The devastating earthquake of 2015 caused significant damage to homes, schools, temples, and infrastructure, but the resilience and determination of the Nepalese people led to remarkable recovery and reconstruction efforts. Nepal celebrates numerous festivals throughout the year, including Dashain, Tihar, Holi, Buddha Jayanti, and Indra Jatra, reflecting the country's multicultural and multi-religious society. The Nepalese people are widely recognized for their hospitality, kindness, and strong sense of community. With its spectacular landscapes, diverse wildlife, rich cultural heritage, and warm-hearted people, Nepal continues to inspire travelers, researchers, and nature lovers from across the globe, making it a unique destination that combines adventure, spirituality, and tradition with modern aspirations for sustainable development and economic growth.
"""

# Add the T5 task prefix
input_text = "summarize: " + text

# Tokenize the input
inputs = tokenizer.encode(
    input_text,
    return_tensors="pt",
    max_length=512,
    truncation=True
)

# Generate the summary
summary_ids = model.generate(
    inputs,
    max_length=100,
    min_length=30,
    length_penalty=2.0,
    num_beams=4,
    early_stopping=True
)

# Decode the summary
summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)

# Print the result
print("Original Text:\n")
print(text)

print("\n" + "="*60)

print("\nSummary:\n")
print(summary)
```
